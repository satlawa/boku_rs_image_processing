```python
#path = "/media/philipp/ed7d22ba-5a3b-4d31-bf6c-6add6e106b3d/crops/s1_raster/march/gee_training_S1_org.csv"
#path = "/media/philipp/ed7d22ba-5a3b-4d31-bf6c-6add6e106b3d/crops/s1_raster/march/gee_training_S1_ext.csv"
path = "/media/philipp/ed7d22ba-5a3b-4d31-bf6c-6add6e106b3d/crops/s1_raster/march/gee_testing_S1.csv"
data_org = pd.read_csv(path)
```

```python
data_org = data_org.drop('.geo', axis=1)
```

```python
data_org['system:index'] = data_org['system:index'].astype('str')
# extract date
data_org['date'] = data_org['system:index'].str[17:25]
data_org['poi'] = data_org['system:index'].str[85:-2]
```

```python
scenes = data_org['date'].unique()
```

```python
data_org
```

```python
s1_list = []
set_crop_id = 0
for scene in scenes:
    # filter dataframe
    temp = data_org.loc[data_org['date']==scene,['id', 'Crop_ID_in', 'VH', 'VV']]
    print(temp.shape)
    # add just tables with all pois
    if temp.shape[0] == 4584: #train_org 1200 #train_ext 23930 #test 4584
        temp = temp.set_index('id')
        if set_crop_id == 0:
            temp.rename(columns={'Crop_ID_in': 'crop_id'}, inplace=True)
            s1_list.append(temp['crop_id'].astype(np.int8))
            set_crop_id = 1
        temp = temp[['VH', 'VV']]
        temp.columns = ['VH_'+scene, 'VV_'+scene]
        #temp = temp.reset_index(drop=True)
        s1_list.append(temp)
```

```python
result = pd.concat(s1_list, axis=1)
```
