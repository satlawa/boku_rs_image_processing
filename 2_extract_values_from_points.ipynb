```python
import os
import numpy as np
import pandas as pd
import geopandas as gpd
import rasterio
```

```python
######################
###   production   ###
######################
```

```python
def extract_values(poi, dir_rast):
    
    # extract coordinates from vector
    coords = [(x,y) for x, y in zip(poi.geometry.x, poi.geometry.y)]
    
    # get sorted file names from raster directory
    files = sorted(os.listdir(dir_rast))
    
    # loop over all files
    for file in files:
        if file == 'DEM_10.jp2':
            next
        else:
            # open raster
            raster = rasterio.open(dir_rast+file)
            # get raster values for coordinates
            poi[file[:file.find('_')] + '_' + dir_rast[-5:-1]] = [x[0] for x in raster.sample(coords)]
            # close raster
            raster.close()
        
    return poi
```
